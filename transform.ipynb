### Transform 함수 이해하기

- 원래의 index를 유지한채 통계 함수를 적용할 수 있음
- groupby와 사용해서 전체 그룹의 통계가아닌 특정 그룹의 통계값으로 계산해서 추가할 수 있음!

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("../../data/titanic_train.csv")
```

```python
data.head()
```

- Pclass에 따른 평균 나이를 추가

```python
data['Age_mean'] = data.groupby('Pclass')['Age'].transform(np.mean)
```

```python
data.head()
```

0번째 row는 Pclass 3인 사람들의 평균 age값을 age_mean값으로 가짐

1번째 row는 Pclass 1인 사람들의 평균 age값을 age_mean값으로 가짐
